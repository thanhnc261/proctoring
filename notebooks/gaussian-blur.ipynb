```python
import cv2 as cv
import numpy as np
```

```python
cap = cv.VideoCapture(0)
if cap.isOpened():
    prev_blur = None

    while True:
        ok, frame = cap.read()
        if not ok:
            break
        
        gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        curr_blur = cv.GaussianBlur(gray, (21, 21), 0)

        if prev_blur is not None:
            diff = cv.absdiff(curr_blur, prev_blur)
            
            _, diff_thresh = cv.threshold(diff, 10, 255, cv.THRESH_BINARY)
            
            cv.imshow("1. Original (Camera)", frame)
            cv.imshow("2. Gaussian Blur (21x21)", curr_blur)
            cv.imshow("3. Motion Detected (After Blur)", diff_thresh)

        prev_blur = curr_blur.copy()

        if cv.waitKey(1) & 0xFF == ord("q"):
            break

    cap.release()
    cv.destroyAllWindows()
```
